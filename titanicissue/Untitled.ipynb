```python
titanic_data.drop("body",axis = 1,inplace = True)
```

```python
titanic_data.head()
```

```python
titanic_data.dropna(inplace = True)  # removes null values
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data.head()  # logistic regression takes 2 values
```

```python
Sex = pd.get_dummies(titanic_data['sex'],drop_first=True)
```

```python
embark = pd.get_dummies(titanic_data["embarked"],drop_first = True)
embark.head(5)
```

```python
pcl = pd.get_dummies(titanic_data["pclass"],drop_first = True)
embark.head(5)
```

```python
titanic_data = pd.concat([titanic_data,Sex,embark,pcl],axis = 1)  #combines the tbale
```

```python
titanic_data.head()
```

```python
titanic_data.drop(['sex','embarked','pclass','home.dest','name','ticket','cabin'],axis = 1,inplace = True)
```

```python
titanic_data.head()
```

```python
titanic_data.drop(['boat'],axis = 1,inplace = True)
```

```python
titanic_data.head()
```

train data


```python
x = titanic_data.drop("survived",axis =1)
```

```python
y = titanic_data["survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train ,X_test,Y_train,Y_test = train_test_split(x,y,test_size =0.3,random_state = 1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train,Y_train)
```

```python
prediction = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(Y_test,prediction)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(Y_test,prediction)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,prediction)
```
